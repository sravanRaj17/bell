```python
from qiskit import QuantumCircuit,ClassicalRegister,QuantumRegister
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
eb1=1
eb2=1
a_bit=0
a_basis=0
b_basis=0
```

```python
Q=QuantumRegister(1,"Q")
A=QuantumRegister(1,"A")

c1=ClassicalRegister(1,"c1")
c2=ClassicalRegister(1,"c2")
a_qc=QuantumCircuit(Q,A,c1,c2)
display(a_qc.draw(output="mpl"))
```

```python

if (a_bit==1):
    a_qc.x(Q)
if (a_basis==1):
    a_qc.h(Q)
if (a_basis==1):
    a_qc.h(Q)
    a_qc.cx(Q,A)
    a_qc.h(Q)

a_qc.h(Q)

a_qc.measure(Q,c1)
a_qc.measure(A,c2)

display(a_qc.draw(output="mpl"))
```

```python
result=AerSimulator().run(a_qc,shots=100).result()
stats=result.get_counts()
print(stats)
```

```python
key_string=list(stats.keys())[0]
m2,m1=map(int, key_string.split())
print(f"m1={m1} , m2={m2}")
```

```python
B=QuantumRegister(1,"B")
f=ClassicalRegister(1,"f")
a_qc=QuantumCircuit(B,f)
if (eb2==0):
    a_qc.h(B)
if (eb2==1):
    a_qc.x(B)
    a_qc.h(B)
    

    
if (m1==1):
    a_qc.x(B)
if (m2==1):
    a_qc.z(B)
if (b_basis==1):
    a_qc.h(B)
if (eb2==1 and b_basis==1 and m1==0 and m2==0):
    a_qc.x(B)
if (eb2==1 and b_basis==1 and m1==0 and m2==1):
    a_qc.x(B)
if (eb2==1 and b_basis==1 and m1==1 and m2==1):
    a_qc.x(B)
a_qc.measure(B,f)

display(a_qc.draw(output="mpl"))
```

```python
result=AerSimulator().run(a_qc,shots=100).result()
stats=result.get_counts()
print(stats)
```
